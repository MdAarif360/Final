```python
import numpy as np
from joblib import load

# Assuming 'pt' and 'similarity_score' are saved using joblib
pt = load('cleaned_dataframe.joblib')
similarity_score = load('similarity_score.joblib')

def recommend(movie_title):
    # Strip whitespace from the index values
    pt.index = pt.index.str.strip()

    # Check if the movie title is in the cleaned index
    if movie_title not in pt.index:
        print(f"Movie title '{movie_title}' not found in the index.")
        return

    # Fetch the index
    index = np.where(pt.index == movie_title)[0][0]

    # Check if the index is out of bounds
    if index >= len(similarity_score):
        print(f"Index {index} is out of bounds for similarity_score.")
        return

    # Get similar items
    similar_items = sorted(enumerate(similarity_score[index]), key=lambda x: x[1], reverse=True)[1:6]

    # Print similar movie titles
    for i in similar_items:
        print(pt.index[i[0]])

# Example usage:
recommend('Zodiac')
```
